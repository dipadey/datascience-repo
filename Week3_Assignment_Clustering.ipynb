```python
def get_category_type(row):
    try:
        categories_list=row['categories']
    except:
        categories_list=row['venue.categories']
    if len(categories_list)==0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues=results['response']['groups'][0]['items']
nearby_venues=json_normalize(venues)
filtered_columns=['venue.name','venue.categories','venue.location.list','venue.location.lng']
nearby_venues=nearby_venues.loc[:,filtered_columns]
nearby_venues['venue.categories']=nearby_venues.apply(get_category_type,axis=1)
nearby_venues.columns=[col.split(".")[-1] for col in nearby_venues.columns]
nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    venues_list=[]
    for name,lat,lng in zip(names, latitudes, longitudes):
        print (name)
    url='https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        lat,
        lng,
        radius,
        LIMIT)
    results=requests.get(url).json()["response"]['groups'][0]['items']
    venues_list.append([(
        name,
        lat,
        lng,
        v['venue']['name'],
        v['venue']['location']['lat'],
        v['venue']['location']['lng'],
        v['venue']['categories'][0]['name']) for v in results])
    nearby_venues=pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns=['Neighborhood',
                          'Neighborhood Latitude',
                          'Neighborhood Longitute',
                          'Venue',
                          'Venue Latitude',
                          'Venue Longitute',
                          'Venue Category']
    return (nearby_venues)
```

```python
#dftr.head()
dftr_venues=getNearbyVenues(names=dftr['Neighbourhood'],
                           latitudes=dftr['Latitude'], 
                           longitudes=dftr['Longitude'])
```

```python
dftr_venues.groupby('Neighborhood').count()
```

```python
dftr_onehot=pd.get_dummies(dftr_venues[['Venue Category']], prefix="",prefix_sep="")
dftr_onehot['Neighborhood']=dftr_venues['Neighborhood']
fixed_columns=[dftr_onehot.columns[-1]] + list(dftr_onehot.columns[:-1])
dftr_onehot=dftr_onehot[fixed_columns]
dftr_onehot.head()
```

```python
dftr_grouped=dftr_onehot.groupby('Neighborhood').mean().reset_index()
dftr_grouped
```

```python
kclusters=1
dftr_grouped_clustering=dftr_grouped.drop('Neighborhood',1)
kmeans=KMeans(n_clusters=kclusters, random_state=0).fit(dftr_grouped_clustering)
kmeans.labels_[0:10]
```
